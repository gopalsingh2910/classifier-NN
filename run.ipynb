```python
## load and process images

import numpy as np
import src.config as config
import matplotlib.pyplot as plt
from src.dataset import load_training_data, load_testing_data

X_train, y_train = load_training_data()
X_test, y_test = load_testing_data()

choice = np.random.choice(X_train.shape[0], 12, replace=False)
sample_train_images = X_train[choice]
sample_train_labels = y_train[choice]

plt.figure(figsize=(6, 6))
for i in range(sample_train_images.shape[0]):
    plt.subplot(3, 4, i+1)
    plt.imshow((sample_train_images[i].reshape(config.input_dim, config.input_dim, 3)*255).astype('uint8'))
    plt.title(f"Label: {sample_train_labels[i]}")

    plt.tight_layout()
    plt.axis('off')
plt.show()
```

```python
from src.model import Model

model = Model()
model.train(X_train, y_train)
accuracy = model.test(X_train, y_train)
print(f'Training Accuracy: {accuracy * 100:.2f}%')
accuracy = model.test(X_test, y_test)
print(f'Test Accuracy: {accuracy * 100:.2f}%')
```
